```python
initial_state = np.matrix([[0.2, 0.8]])
transition_matrix = np.matrix([[0.9, 0.1], [0.7, 0.3]])
```

```python
print('initial state:')
print(initial_state)
```

```python
print('transition probability matrix: ')
print(transition_matrix)
```

```python
def state_transition(state_n, n):
    for i in range(n):
        state_n = np.dot(state_n, transition_matrix)
    return state_n
```

```python
# one week in
state_1 = np.dot(initial_state, transition_matrix)
state_1
```

```python
# two weeks in
state_2 = np.dot(state_1, transition_matrix)
state_2
```

```python
# three weeks in
state_3 = np.dot(state_2, transition_matrix)
state_3
```

```python
# ten weeks in?
state_transition(initial_state, 10)
```

```python
# how about 20 weeks in?
state_transition(initial_state, 20)
```

```python
# how about 200 weeks in?
state_transition(initial_state, 200)
```

#### seems like we have reached a state, where the marketshare is no longer changing 
#### we refer to this state as stationary Matrix, and the system is said to be at steady state
