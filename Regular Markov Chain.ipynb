# Regular Markov Chains 



A transition matrix P is regular if some power of P has only positive entries. <br /> 
A markov chain is a regular markov chain if its transition matrix is regular.

### Properties of regular markov chain:<br />
    1.There is a unique stationary matrix S that can be found by solving the equation 
                                        S*P = S          (* denotes dot)
    2.Given any initial state matrix S_0, the state matrices approcach the stationary matrix S 
    3.The matrices p_k approach a limiting matrix p, where eah row of p is equal to the stationary matrix S
